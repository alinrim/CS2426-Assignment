# Part 2 – Statistical Analysis


## Pattern 1 – investigating average

I want to investigate which job title yields the highest average salary

```python
salary_per_job = df.groupby('job_title')['salary'].agg(['min','max', 'mean', 'median', 'std']).sort_values('mean', ascending = False)
salary_per_job
```

```python
degree_counts = df.groupby(['job_title', 'education_level']).size().unstack(fill_value=0)
degree_counts
```

**Write here your statistical discussion of Pattern 1.**

AI Engineer and Machine Learning Engineer roles have the highest average salaries (140k), while Data Analyst, Data Scientist, and Business Analyst roles are lower (99–101k). Most technical roles tend to require a Master’s or PhD, whereas business-focused roles have a more even mix of education levels.

This shows a clear link between role complexity, education, and pay: advanced degrees are more common in high-paying, technical positions, while other roles are more flexible.

# Part 3 – Visualizations

## Plot 1

I want to see the average salary over time....

```python
import matplotlib.pyplot as plt
yearly_salary = df.groupby('job_posting_year')['salary'].mean()
plt.figure(figsize = (8,6))
plt.plot(yearly_salary.index, yearly_salary.values, marker = 'o', linestyle = '-')
plt.title('Average Salary Over Time (2020-2026)')
plt.xlabel('Year')
plt.ylabel('Average Salary')
plt.show()
```


**Write here a short comment explaining what Plot 1 shows and why it is interesting.**


The plot shows how the avergae salary for Ai and Data Science job has changed from 2020-2026. There is a clear trend of fluctuation in the salary over time (around 110k - 115k), indicating that this field is constantly evolving.we can see that the average salary peaked the most in 2024 and took a in 2025, which again goes to show that even though the demand for Data Science and AI roles have become quite high, the salaries have remained steady.

## Plot 2

I want to investigate the job distribustion among the years...

```python
job_trends = df.groupby(['job_posting_year', 'job_title']).size().unstack(fill_value=0)
top_jobs = df['job_title'].value_counts().head(4).index
filtered_trends = job_trends[top_jobs]
filtered_trends.plot(kind = 'bar', figsize = (12,6),width = 0.8, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])
plt.title('Hiring Trends: Dominant Roles (2020-2026)', fontweight = 'bold')
plt.xlabel('Year')
plt.ylabel('Number of Job Postings')
plt.legend(title = 'Job Title', bbox_to_anchor = (1,1))
plt.xticks(rotation = 0)
plt.show()
```


**Write here a short comment explaining what Plot 2 shows and why it is interesting.**


The plot shows the number of job postings for the most common roles from 2020 to 2026. Overall, the demand for these roles remains fairly consistent across the years, with no major spikes or drops in hiring.

We can observe that all four roles follow a similar pattern, with only small fluctuations between years. This suggests that demand for these core AI and Data Science roles is stable over time, indicating a steady and mature job market where hiring needs remain consistently high rather than rapidly changing.


## Plot 3


I want to investigate the demand on education levels....

```python
edu_counts = df['education_level'].value_counts()
my_explode = [0] * len(edu_counts)
my_explode[0] = 0.1
plt.figure(figsize = (8,8))
edu_counts.plot.pie(autopct='%1.1f%%',shadow = True,explode = my_explode[:len(my_explode)]
                    ,startangle = 140
                    ,colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
plt.title('Distribution of Education Levels')
plt.show()
```


**Write here a short comment explaining what Plot 3 shows and why it is interesting.**

The pie chart shows the distribution of education levels among AI and Data Science job postings. The three main categories, Bachelor’s, Master’s, and PhD are almost evenly represented, with 32.8% of jobs requiring a Bachelor’s, 34.1% a Master’s, and 33.1% a PhD.

This indicates that employers value advanced education highly across the board, and that there is no single dominant education level for these roles. It suggests that candidates with different levels of formal education can find opportunities in the field, highlighting the diversity of hiring requirements.
